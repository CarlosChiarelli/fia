```r
# bibliotecas
library(tidyverse)
library(lubridate)
library(ggthemes)
library(ggcorrplot)
library(patchwork)

# tema para series temporais
theme_set(theme_economist())

####### FUNÇÕES #######

colunasAlvo <- function(dfCompleto, coluna="close"){
  # pega o df completo e retorna df com 3 colunas
  # data, codigo, coluna escolhida (open, close, high, low, volume, adjusted)
  
  dfCompleto %>% 
  select(
    ends_with(coluna), data
  ) %>% 
  rename_at(
    vars(-data),
    function(col){ str_sub(col, end = 5) }
  ) %>% 
  gather(codigo, !!coluna, -data)
}
  
plotaColAlvo <- function(df, comLegenda = F){
  # plota a coluna alvo escolhida
  
  # pega o nome da coluna a ser plotada
  colAlvo <- names(df)[3]
  
  # faz o grafico
  plot <- df %>% 
    ggplot(
      aes_string("data", colAlvo, col="codigo")
    )+
    geom_line()+
    labs(title = colAlvo)
  
  # decide se quer os rotulos da legenda
  if(comLegenda){
    plot + theme(legend.position = 'bottom')
  }else{
    plot + theme(legend.position = 'none')
  }
}

plotaDuasColunas <- function(df, col1, col2){
  # plota duas colunas principais juntas
  
  g1 <- df %>% 
    colunasAlvo(col1) %>% 
    plotaColAlvo()

  g2 <- df %>% 
    colunasAlvo(col2) %>% 
    plotaColAlvo()

  g1 / g2
}

####### FUNÇÕES #######
```

Outro ponto encontrado é que não existe dados do sábado e do domingo, dias 7 e 1 da semana (respectivamente), pois a B3 não opera nesses dias. Ainda existe missings temporais que não parecem nos dados como podemos observar no gráfico abaixo.

```r
aux <- acoesDf %>% 
  select(
    data, starts_with('petr3'), -petr3_sa_adjusted, -petr3_sa_volume
  ) %>% 
  mutate(
    diaSemana = wday(data)
  ) 

tibble(data = seq(ymd('2017-01-10'),ymd('2020-01-01'), 1)) %>% 
  left_join(aux) %>% 
  mutate(
    nulos = if_else(is.na(diaSemana), 1, 0),
    somaAcumNAs = cumsum(nulos),
  ) %>% 
  select(data, somaAcumNAs, nulos) %>% 
  filter(data<'2017-02-28') %>% 
  ggplot(
    aes(data, somaAcumNAs, col=nulos)
  )+ 
  geom_line()+
  geom_point()+
  labs(title = "Soma acumulada dos missings - 2017")+
  theme(legend.position = 'none')
```

# Correlação

```r
acoesDf %>% 
  select(
    starts_with('petr3'),
    starts_with('vale3'),
    -data
  ) %>% 
  cor() %>% 
  ggcorrplot()
```

Visualizando a correlação das colunas da Petrobras e da Vale vemos que não existe uma relação linear entre elas. Caso adicionássemos indicadores (médias móveis, bandas de bollinger, IFR) poderíamos observar uma certa relação linear entre essas variáveis.
