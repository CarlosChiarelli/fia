# Como R identifica o tamanho dos pontos?

```r
aux <- df %>% 
    select(nota_mt, nota_redaTot, sociEco_renda)
```

```r
head(aux)
```

```r
aux %>% 
    # pegando 50% dos dados
    sample_frac(.15) %>% 
    filter(nota_mt != 0 & nota_redaTot != 0) %>% 
    #filter(!(nota_mt != 0 & nota_redaTot != 0)) %>% 
    ggplot(
        aes(nota_mt, nota_redaTot, size = sociEco_renda, col = sociEco_renda)
    )+
    geom_jitter(alpha = .5, height = 9)+
    geom_vline(xintercept = 600, lty = 2, size = .8)+
    geom_hline(yintercept = 600, lty = 2, size = .8)+
    scale_color_viridis_c(direction = -1)+
    labs(title = 'Nota Redação x Nota Matemática - RENDA\n')+
    theme_light()
    
```

## Parâmetro **Dodge** do **ggplot2**

```r
aux <- df %>% 
    count(estado, sexo) %>% 
    arrange(desc(n))

maioresEst <- aux %>% 
    .$estado %>% 
    unique()
```

```r
head(aux,20)
```

```r
aux %>% 
    mutate(estado = fct_relevel(estado, maioresEst)) %>% 
    ggplot(aes(estado, n, fill=sexo))+
    geom_bar(
        stat = 'identity'
    )
```

```r
aux %>% 
    mutate(estado = fct_relevel(estado, maioresEst)) %>% 
    ggplot(aes(estado, n, fill=sexo))+
    geom_bar(
        stat = 'identity',
        position = 'dodge',
        width=.7
    )
```

```r
aux %>% 
    mutate(estado = fct_relevel(estado, maioresEst)) %>% 
    ggplot(aes(estado, n, fill=sexo))+
    geom_bar(
        stat = 'identity',
        position = 'fill',
        width=.7
    )
```

## Funções **spread, unite, separate**

```r
df <- tibble(mes=rep(c('jan','fev','marc'),2),
                 nome=rep(c("Amy", "Bob"), each=3),
                 nota_mat=c(9, 7, 6, 8, 6, 9),
                 nota_bio=c(6, 7, 8, 5, 6, 7))
```

```r
head(df)
```

```r
aux <- df %>% 
    select(-nota_bio)
```

```r
head(aux)
```

```r
aux <- aux %>% 
    spread(nome, nota_mat)

head(aux)
```

```r
aux <- df %>% 
    mutate(nota = str_c(nota_mat, '-',nota_bio)) %>% 
    select(-nota_mat, -nota_bio)
```

```r
aux
```

```r
aux <- aux %>% 
    separate(nota, into = c("nota1", "nota2"), sep = '-')

head(aux)
```

```r
aux %>% 
    unite(novaCol, nota1, nota2, sep='-')
```
