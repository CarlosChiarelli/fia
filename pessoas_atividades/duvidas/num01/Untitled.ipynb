```r
aux <- df %>% 
    count(estado, sexo) %>% 
    arrange(desc(n))

maioresEst <- aux %>% 
    .$estado %>% 
    unique()
```

```r
aux %>% 
    mutate(estado = fct_relevel(estado, maioresEst)) %>% 
    ggplot(aes(estado, n, fill = sexo))+
    geom_bar(stat = 'identity')+
    scale_y_continuous(
        breaks = seq(0, 2000, 200),
        labels = seq(0, 2000, 200)
    )
```

```r
aux %>% 
    mutate(estado = fct_relevel(estado, maioresEst)) %>% 
    ggplot(aes(estado, n, fill = sexo))+
    geom_bar(
        stat = 'identity',
        position = 'dodge',
        width = .7
    )+
    scale_y_continuous(
        breaks = seq(0, 2000, 200),
        labels = seq(0, 2000, 200)
    )
```

```r
aux %>% 
    mutate(estado = fct_relevel(estado, maioresEst)) %>% 
    ggplot(
        aes(estado, n, fill = sexo)
    )+
    geom_bar(
        stat = 'identity',
        position = 'fill'
    )
```
